```python
# --1
crime_data = pd.read_csv("crime_data.csv")
crime_data.head()
```

```python
# --2
crime_data_cleaned = crime_data.copy()
crime_data_cleaned = crime_data_cleaned.loc[:, (crime_data_cleaned.isna().sum() /len(crime_data_cleaned)) <= 0.5]
print(crime_data_cleaned)
```

```python
# --3
crime_data_cleaned["DATE OCC"] = pd.to_datetime(crime_data_cleaned["DATE OCC"], format='%m/%d/%Y %I:%M:%S %p')

crime_data_cleaned["Year"] = crime_data_cleaned["DATE OCC"].dt.year
crime_data_cleaned["Month"] = crime_data_cleaned["DATE OCC"].dt.month
crime_data_cleaned["Day"] = crime_data_cleaned["DATE OCC"].dt.day

crime_data_cleaned["TIME OCC"] = crime_data_cleaned["TIME OCC"].astype(str).str.zfill(4) # make sure it is in a 4 digit format
crime_data_cleaned["HOUR"] = pd.to_datetime(crime_data_cleaned["TIME OCC"], format='%H%M').dt.time

crime_data_cleaned
```

```python
# --4
crime_data_cleaned2 = crime_data_cleaned[crime_data_cleaned['Year'] == 2023]

crime_data_cleaned3 = crime_data_cleaned2[crime_data_cleaned2['Crm Cd Desc'] == 'BURGLARY']

crime_data_cleaned3
```

```python
# --5
crime_data_cleaned.groupby("AREA NAME")["Vict Age"].agg(['size', "mean"]).sort_values('size', ascending=False)
```

```python
crime_data2 = crime_data.copy()

def calculate_severity(x):
    score = 0
    if not pd.isna(x["Weapon Used Cd"]):
        score+=5
    if x["Crm Cd Desc"] == "BURGLARY":
        score+=3
    if score == 0:
        score+=1
    return score


crime_data2["Severity Score"] =  crime_data2.apply(calculate_severity, axis=1)

crime_data2.groupby("AREA NAME")["Severity Score"].sum()
```
